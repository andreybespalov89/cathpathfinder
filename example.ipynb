# С валидатором

```python
def custom_validator(chain):
    random_float = random.random()
    if (random_float > 0.6):
        return False
    else:
        return True

algo = AlgoSeqDynamicWithValidation(custom_validator)
a_seq = "ACCDACCDDD"
b_seq = "CBBBBFFF"
steps, operations, distance = algo.run(a_seq, b_seq)

if steps is None:
    print("Не найден валидный путь преобразования")
else:
    print(f"Найден путь! Расстояние: {distance}")
    print(f"Операции: {operations}")
    print(f"Шаги преобразования: {steps}")
    print("Последовательность преобразования:")
    for i, step in enumerate(steps):
        print(f"{i}: {step}")
```
