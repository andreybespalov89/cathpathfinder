```python
import random
print('pre import c lib')
import datasetgradualstep_c as dsc
print('post import c lib')
pairs_path = "layer_outputs/pairs_1_layer.tsv"
fasta_path = "layer_outputs/pdb_sequences_layer_1.fasta"
print('read fasta')
# читаем FASTA в массив
seqs = []
with open(fasta_path, "r", encoding="utf-8") as f:
  cur = None
  for line in f:
      if line.startswith(">"):
          if cur is not None:
              seqs.append(cur)
          cur = ""
      else:
          cur += line.strip()
  if cur is not None:
      seqs.append(cur)
print('readed fasta')
# читаем все пары и выбираем случайную
pairs = []
print('reading file')
with open(pairs_path, "r", encoding="utf-8") as f:
  for line in f:
      if not line.strip():
          continue
      cols = line.rstrip("\n").split("\t")
      if len(cols) < 4:
          continue
      i1 = int(cols[2])
      i2 = int(cols[3])
      pairs.append((i1, i2))
print('readed file')
if not pairs:
  raise RuntimeError("Файл пар пуст")
print('random choice')
idx1, idx2 = random.choice(pairs)
a_seq = seqs[idx1]
b_seq = seqs[idx2]

print(f"Выбрана пара idx1={idx1} idx2={idx2}")

def custom_validator(chain):
  return random.random() <= 0.6
print('Start')
steps, operations, distance = dsc.algo_seq_dynamic_with_validation_run(
  a_seq, b_seq, validator=custom_validator
)
print('Completed')

if steps is None:
  print("Не найден валидный путь преобразования")
else:
  print(f"Найден путь! Расстояние: {distance}")
  print(f"Операции: {operations}")
  print(f"Шаги преобразования: {steps}")
  print("Последовательность преобразования:")
  for i, step in enumerate(steps):
      print(f"{i}: {step}")
```
